```python
from sklearn.metrics import mean_squared_error

reals_matrix = frame_of_beers.as_matrix()

reals_matrix.shape
```

```python
def my_squared_error(reals_matrix,approx):
    sum_squared_error = 0
    ratings_count = 0
    for j in range(reals_matrix.shape[0]):
        for n in range(reals_matrix.shape[1]):
            if np.isnan(reals_matrix[j][n]) == False:
                sum_squared_error = (approx[j][n] - reals_matrix[j][n])**2 + sum_squared_error
                ratings_count = ratings_count + 1
    mean_square_error = sum_squared_error/ratings_count
    return mean_square_error

my_squared_error(reals_matrix,approx)
```

```python
k_errors_list = []
for i in range(1,100):
    [A,B,C] = svds(frame_of_beers, i)
    predicted_ys = np.dot(A,C)
    k_error = my_squared_error(reals_matrix,predicted_ys)
    k_errors_list.append(k_error)

k_errors_list
```
